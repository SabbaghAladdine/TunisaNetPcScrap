```python
import requests
import re
base_url = "https://www.tunisianet.com.tn/702-ordinateur-portable"
```

```python
response = requests.get(base_url)
```

```python
response.ok
```

```python
from bs4 import BeautifulSoup
```

```python
data = response.content
```

```python
soup = BeautifulSoup(data)
```

```python
pr = soup.find_all("div",class_="item-product")
len(pr)
```

```python
Products = {}
```

```python
tp = soup.find("div",class_="total-products").text
print(re.findall(r'\d+',tp))
t=int((re.findall(r'\d+',tp)[0]))
pages= int(t/24)
print(pages)
len(Products)
```

```python
for x in range(1,pages+2):
    print(x)
    response = requests.get(base_url+"?page="+str(x)+"&order=product.price.asc")
    soup=BeautifulSoup(response.content)
    pr = soup.find_all("div",class_="item-product")
    n=1
    for p in pr:
        Products[p.find("h2").text+" "+str(n)] = [p.find("div",itemprop="description").text[:-78],p.find("span",class_="price").text]
        print(str(n)+"added "+p.find("h2").text)
        n=n+1
    print(len(Products))    
```
